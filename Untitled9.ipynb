```python
import pandas as pd
import numpy as np
import datetime
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Creating DataSets
```

```python
Crimes1= pd.read_csv('Chicago_Crimes_2008_to_2011.csv',error_bad_lines=False)
```

```python
Crimes2= pd.read_csv('Chicago_Crimes_2012_to_2017.csv',error_bad_lines=False)
```

```python
Crimes=pd.concat([Crimes1, Crimes2])
```

```python
Crimes.shape
```

```python
del Crimes1
del Crimes2
```

```python
Crimes.drop_duplicates(subset=['ID', 'Case Number'], inplace=True)
```

```python
Crimes.shape
```

```python
Crimes.head(3)
```

```python
Crimes.Date = pd.to_datetime(Crimes.Date, format='%m/%d/%Y %I:%M:%S %p')
```

```python
Crimes.sort_values(by='Date', inplace=True)
```

```python
Crimes.index = pd.DatetimeIndex(Crimes.Date)
```

```python
Crimes.info()
```

```python
#Data Cleaning for Outliers
```

```python
Crimes.drop(['Unnamed: 0', 'Case Number', 'IUCR','Updated On','Year', 'FBI Code',
             'Beat','Ward', 'Location'], inplace=True, axis=1)
```

```python
Crimes.head()
```

```python
loc_desc_to_change  = list(Crimes['Location Description'].value_counts()[20:].index)
desc_to_change = list(Crimes['Description'].value_counts()[20:].index)
```

```python
Crimes.loc[Crimes['Location Description'].isin(loc_desc_to_change) , Crimes.columns=='Location Description'] = 'OTHER'
Crimes.loc[Crimes['Description'].isin(desc_to_change) , Crimes.columns=='Description'] = 'OTHER'
```

```python
Crimes['Primary Type']         = pd.Categorical(Crimes['Primary Type'])
Crimes['Location Description'] = pd.Categorical(Crimes['Location Description'])
Crimes['Description']          = pd.Categorical(Crimes['Description'])
```

```python
#Analysis Begins
```

```python
sns.displot(Crimes['Date'],aspect=3)
```

```python
#Analysis by Location of Crimes
```

```python
plt.figure(figsize=(15,6))
sns.countplot(y='Location Description',data= Crimes,hue='Arrest', order= Crimes['Location Description'].value_counts().index)
```

```python
#Analysis by Type of Crimes
```

```python
plt.figure(figsize=(8,10))
sns.countplot(y='Primary Type',data= Crimes,orient= 'h', order= Crimes['Primary Type'].value_counts().index)
```

```python
Crimes_Count_Date = Crimes.pivot_table('ID', aggfunc=np.size, columns='Primary Type', index=Crimes.index.date, fill_value=0)
Crimes_Count_Date.index = pd.DatetimeIndex(Crimes_Count_Date.index)
Crimes_Count_Date
```

```python
plo = Crimes_Count_Date.rolling(365).sum().plot(figsize=(15, 30),subplots=True, layout=(-1, 2), sharex=False, sharey=False)
```

```python
#ANALYSIS BY DAY AND TIME
```

```python
Crimes['Hour'] = Crimes['Date'].apply(lambda time: time.hour)
Crimes['Month'] = Crimes['Date'].apply(lambda time: time.month)
Crimes['Day of Week'] =Crimes['Date'].apply(lambda time: time.dayofweek)
dmap= {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
Crimes['Day of Week'] = Crimes['Day of Week'].map(dmap)
```

```python
plt.figure(figsize=(6,6))
sns.countplot(y='Day of Week',data=Crimes,order= Crimes['Day of Week'].value_counts().index)
```

```python
dayHour = Crimes.groupby(by=['Day of Week','Hour']).count()['Primary Type'].unstack()
```

```python
dayHour.head()
```

```python
#Heatmap for Crimes Based on Day of the Week and Hour of the Day
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(dayHour)
```

```python
#MAJOR CRIME TYPES AND THEIR LOCATIONS
```

```python
Crimes_new = Crimes[(Crimes['Primary Type'] == 'SEX OFFENSE') | (Crimes['Primary Type'] == 'HOMICIDE') | 
                    (Crimes['Primary Type'] == 'ARSON')]
ax = sns.lmplot('Longitude', 'Latitude',
                data= Crimes_new[['Longitude','Latitude']],
                fit_reg=False,
                height=4, 
                scatter_kws={'alpha':.1})
ax = sns.kdeplot(data=Crimes_new[['Longitude','Latitude']],
                palette='coolwarm',
                 bw_method=.005,
                 cbar=True, 
                 shade=False, 
                 thresh=False)
ax.set_xlim(-87.9,-87.5)
ax.set_ylim(41.60,42.05)
```
